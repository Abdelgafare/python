# Data cleaning 

```python
### there is no nulls and we going to drop the empty column 
df=df.drop(columns='Unnamed: 32')
```

```python
### we need to change diagnosis  to make look like a dummy variable 

diagnosis_dummies = pd.get_dummies(df['diagnosis'], prefix='diagnosis', drop_first=True)
diagnosis_dummies = diagnosis_dummies.astype(int)  # Convert dummy variables to integers (0 and 1)
df = pd.concat([df, diagnosis_dummies], axis=1)
df = df.drop(columns=['diagnosis'])  
df.head(10)
```

```python
df['diagnosis_M'].value_counts()
```

# Data modeling

```python
### lets identify my depdent and  indepdents variables 
X=df.drop(columns=['diagnosis_M','id'])
Y=df['diagnosis_M']
```

```python
### normalizing the data because not all have the same unit 
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)

# X_scaled is now a numpy array with normalized data
```

```python
### split the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=0)
```

```python
#Fit  Logistic Regression 
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#predict
predict=model.predict(X_test)
# confusion matrix 
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(Y_test,predict),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])
```

```python
print(classification_report(Y_test,predict))
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)

model_prediction = model.predict(X_test)
print('Logistic Regression accuracy = ', metrics.accuracy_score(model_prediction,Y_test))
```
